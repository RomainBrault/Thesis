```python
import numpy as np
from sklearn.metrics.pairwise import rbf_kernel
import matplotlib.pyplot as plt
from scipy.optimize import fmin_l_bfgs_b
```

```python
N = 1000
X = (2 * np.random.rand(N) - 1).reshape(N, 1)
# y = np.ones(N).reshape(N, 1)
y = np.random.rand(N).reshape(N, 1)
```

```python
plt.scatter(X, y)
plt.show()
```

```python
K = rbf_kernel(X, gamma=1000000.)
# lbda = 2. * 1. * np.mean(y ** 2) / 1 ** 2
lbda = 1.
```

```python
# C = np.dot(np.linalg.pinv(K + lbda * K.shape[0] / 2 * np.eye(N)), y)
o = fmin_l_bfgs_b(lambda C: np.linalg.norm(np.dot(K, C) - y.T) ** 2 / K.shape[0], np.ones(N), approx_grad=True, bounds=((-.1, .1),) * N)
```

```python
np.linalg.norm(np.dot(K, o[0]) - y) ** 2 / K.shape[0]
```
